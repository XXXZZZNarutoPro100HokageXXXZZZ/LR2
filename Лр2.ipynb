## Решите индивидуальное задание согласно вашему варианту (1). 
Распределение инвестиций. Компания выделила 1 000 000 рублей на три
проекта. Первый проект получает в два раза больше, чем второй, а третий
получает на 200 000 рублей больше, чем второй. Сколько денег получит
каждый проект?

Обозначим за y количество денег выделенных на 2 проект , за x и z - первый и третий соответственно

Сумма всех инвестиций должна быть равно 1000000 р. x + y + z = 1000000

Раз первый проект получает в 2 раза больше денег, соотвественно х = 2y,

так как третий проект получает на 200000 р. больше чем второй, последняя подсистема будет иметь вид z =y + 200000 

Итого получим систему следущего вида :

\begin{equation*}
 \begin{cases}
   x + y + z = 1000000, 
   \\
    х = 2y ,
   \\
   z =y + 200000 .
 \end{cases}
\end{equation*}

```python
import numpy as np
# Создадим матрицу коэффициентов и вектор свободных членов
A = np.array([
    [1, 1, 1],    # x + y + z
    [1, -2, 0],    # x - 2y = 0
    [0, -1, 1]     # -y + z = 200 000
])

B = np.array([1000000, 0, 200000])

## Метод Крамера
def cramer_solve(A, B):
    det_A = np.linalg.det(A)
    solutions = []
    for i in range(A.shape[1]):
        Ai = A.copy()
        Ai[:, i] = B
        det_Ai = np.linalg.det(Ai)
        solutions.append(det_Ai / det_A)
    return solutions

cramer_solution = cramer_solve(A, B)
print(f"Решение методом Крамера: {cramer_solution}")

## Матричный метод (A^(-1)*B)
matrix_solution = np.linalg.inv(A) @ B
print(f"Решение матричным методом: {matrix_solution}")

## Linalg.solve
numpy_solution = np.linalg.solve(A, B)
print(f"Решение np.linalg.solve: {numpy_solution}")

## Проверка решений
print("\nСравнение решений:")
print(f"Метод Крамера и np.linalg.solve: {np.allclose(cramer_solution, numpy_solution)}")
print(f"Матричный метод и np.linalg.solve: {np.allclose(matrix_solution, numpy_solution)}")

## Интерпретация результатов
x, y, z = numpy_solution
print("\nРаспределение инвестиций:")
print(f"Первый проект (x): {x:.2f} руб.")
print(f"Второй проект (y): {y:.2f} руб.")
print(f"Третий проект (z): {z:.2f} руб.")
```
